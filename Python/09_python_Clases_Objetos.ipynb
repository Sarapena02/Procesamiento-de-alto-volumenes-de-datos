### La clase matplotlib.patches.Ellipse, permite dibujar elipses desde matplotlib. La clase se importa de la siguiente manera:

from matplotlib.patchs import Ellipse

Se puede dibujar entonces cualquier elipse, dando los argumentos:

plt.gca().add_patch(Ellipse(xy, width, height, facecolor = color, edgecolor = edgecolor))

Se requiere crear una clase Elipse, la cual el constructor sea con valores iniciales cualesquiera. Adicionalmente, la clase debe tener los siguientes métodos:

1.- Cambiar el ancho

2.- Cambiar el alto

3.- Cambiar el color de relleno

4.- Cambiar el color de borde

5.- Dibujar

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse as mplEllipse

class Elipse:
    def __init__(self, xy, width, height, facecolor='blue', edgecolor='black'):
        """
        Constructor para la clase Elipse.

        Args:
            xy (tuple): Coordenadas del centro de la elipse (x, y).
            width (float): Ancho de la elipse.
            height (float): Alto de la elipse.
            facecolor (str): Color de relleno de la elipse.
            edgecolor (str): Color del borde de la elipse.
        """
        self.xy = xy
        self.width = width
        self.height = height
        self.facecolor = facecolor
        self.edgecolor = edgecolor
        self.patch = mplEllipse(xy, width, height, facecolor=facecolor, edgecolor=edgecolor)
    
    def cambiar_ancho(self, nuevo_ancho):
        """
        Cambia el ancho de la elipse.

        Args:
            nuevo_ancho (float): Nuevo ancho de la elipse.
        """
        self.width = nuevo_ancho
        self.patch.set_width(nuevo_ancho)
    
    def cambiar_alto(self, nuevo_alto):
        """
        Cambia el alto de la elipse.

        Args:
            nuevo_alto (float): Nuevo alto de la elipse.
        """
        self.height = nuevo_alto
        self.patch.set_height(nuevo_alto)
    
    def cambiar_color_relleno(self, nuevo_color):
        """
        Cambia el color de relleno de la elipse.

        Args:
            nuevo_color (str): Nuevo color de relleno.
        """
        self.facecolor = nuevo_color
        self.patch.set_facecolor(nuevo_color)
    
    def cambiar_color_borde(self, nuevo_color):
        """
        Cambia el color del borde de la elipse.

        Args:
            nuevo_color (str): Nuevo color del borde.
        """
        self.edgecolor = nuevo_color
        self.patch.set_edgecolor(nuevo_color)
    
    def dibujar(self, ax):
        """
        Dibuja la elipse en el gráfico dado.

        Args:
            ax (matplotlib.axes.Axes): El objeto de ejes donde se dibujará la elipse.
        """
        ax.add_patch(self.patch)
        ax.autoscale_view()

# Ejemplo de uso
fig, ax = plt.subplots()

# Crear una elipse
mi_elipse = Elipse(xy=(0.5, 0.5), width=0.4, height=0.2, facecolor='red', edgecolor='green')

# Dibujar la elipse
mi_elipse.dibujar(ax)

# Cambiar propiedades
mi_elipse.cambiar_ancho(0.6)
mi_elipse.cambiar_color_relleno('blue')

# Dibujar la elipse con nuevas propiedades
mi_elipse.dibujar(ax)

# Configurar los límites del gráfico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')

plt.show()
```
